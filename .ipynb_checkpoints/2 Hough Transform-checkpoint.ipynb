```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
#%matplotlib inline
```

```python
# Загружаю изображение
image = cv.imread('road1.png')
print(f"size image:{image.shape}")
plt.imshow(image)
```

```python
# Перевожу цветовое пространство из RGB в оттенки серого
gray_img = cv.cvtColor(image, cv.COLOR_RGB2GRAY)
#plt.imshow(gray_img, cmap='gray')
```

```python
# Производим предобработку изображения
# Удаляю шумы c помощью фильтра Гаусса с ядром 5х5
kernel = np.ones((5,5),np.float32)/25
filter_img = cv.filter2D(gray_img, -1, kernel)
#plt.imshow(filter_img, cmap='gray')
```

```python
# По методу Canny выделяю границы (изобр, нижний порог яркости, высший порог яркости)
edge = cv.Canny(filter_img, 50, 100)
plt.imshow(edge, cmap='gray')
```

<h3>Реализую Алгоритм Хафа (Hough Transform)</h3>
Создаю аккумулятор (это нулевая матрица размером с $\rho$ и $\theta$). И параметры настраивающие вычисления:

```python
max_angle = 360//1
k = 1 # уменьшаем шаг угла
pk, lk = 4, 1 # Множитель для акб и множитель для увеличения точности рассчета ro
print(f"Угол:{max_angle} Вциклеугол:{max_angle*k} Шаг:{1/k} ВмассивеАКБячеек:{max_angle*k*pk*lk*max(image.shape[0],image.shape[1])}")
```

```python
accum = np.zeros((max(image.shape[0],image.shape[1])*pk*lk, max_angle*k), float) # градусы, ro
voices = 200 # количество голосов 
```

Функция для вычисления $\rho=x*cos(\theta)+y*sin(\theta)$, где $x,y$ - это координаты пикселя через которую проходит перпендикуляр (координаты пикселя в общем). Дан пример использования функции

```python
def ro(x, y, theta):
    return x*np.sin(theta*np.pi/180)+y*np.cos(theta*np.pi/180)
print(ro(300, 280, 100))
```

Перехожу по каждому пикселю на картинке, нахожу перпендикуляр $\rho$ к линии в этом пикселе от начала координат $O(x_0, y_0)$ под углом $\theta$ и поворачиваю этот угол до *max_angle*.

```python
points = []
tht = [i for i in range(max_angle)]
for x in range(0, len(edge), 1):   
    for y in range(0, len(edge[x]), 1):
        if edge[x][y] != 0: # Отсеиваю нулевые значения (где значение пикселя равно черному цвету)
            pts = []
            for theta in range(0, max_angle*k, 1):
                p = ro(x, y, theta/k)*lk
                pts.append(p) # одна синусойда
                accum[int(p)][theta] += 1
            points.append(pts)
print(max([max(m)/lk for m in points])) # для оценки максимального ro

%matplotlib inline
plt.pcolor(accum, cmap=plt.get_cmap('gist_gray', 11))
plt.colorbar()
plt.show()
```

Учтем голоса в аккумуляторе

```python
# убираем все что меньше порога
# убираем все что меньше порога
def func_for_porog(ac, porog):
    s = []
    for i in range(len(ac)):
        for j in range(len(ac[0])):
            if ac[i][j] == 0:
                pass
            else: 
                if ac[i][j] <= porog:
                    ac[i][j] = 0
                else:
                    s.append([ac[i][j], i, j]) # для отслеживания линий которые выше порога

    return ac, s
accum, max_line = func_for_porog(accum, voices)
```

Построим график фазового пространства Хафа со значениями акуммулятора (светлее-больше значение)

```python
%matplotlib inline
plt.pcolor(accum, cmap=plt.get_cmap('gist_gray', 11))
plt.colorbar()
plt.show()
```

По двум светлым точкам можно понять, что у нас 2 максимума, следовательно, 2 линии на изображении. По этим точкам можно определить параменты угла и полярного радиуса.

Далее построим график косинусов для линии.
Это функция для показа синусов которые получаются. В одном источнике сказано:
Если эти линии (кривые), соответствующие двум точкам,накладываются друг на друга,
 то точка (в пространстве Хафа), где они пересекаются, соответствуетпрямым 
 (в оригинальном месте изображения), которые проходят через обе точки. 
 В общем случае, рядточек, которые формируют прямую линию, определяют синусоиды, 
 которые пересекаются в точкепараметров для той линии. Таким образом, проблема обнаружения 
 коллинеарных точек может быть сведена к проблеме обнаружения пересекающихся кривых.

Теперь наложим эти линии на график, чтобы увидеть где мы определили линии. Выразим фунцию $\rho=x*cos(\theta)+y*sin(\theta)$ как $y(x)=(\rho-x*sin(\theta))/(cos(\theta))$ . Тогда можно посторить линию $y(x)$ следующим образом, реализовав функцию c уже известными параметрами rho и theta:

```python
def f_y(x, rho, theta):
    return (rho - x*np.cos(theta*np.pi/180))/np.sin(theta*np.pi/180)
    
x = [i for i in range(image.shape[0])]
y = []
for i in range(len(max_line)):
    rho, theta = max_line[i][1], max_line[i][2]
    temp = [f_y(n, rho, theta) for n in x]
    y.append(temp)
```

```python
houghLinesImage = np.zeros_like(image) # создаем пустое изображение

# Нарисуйте линии, представленные в аккумуляторе, на исходном изображении
def drawhoughLinesOnImage(image, lines):
    x = [i for i in range(max(image.shape[0],image.shape[1])*pk*lk)]
    y = []
    for i in range(len(lines)):
        rho, theta = lines[i][1], lines[i][2]
        temp = [f_y(n, rho, theta) for n in x]
        y.append(temp)

        #берем для линии две точки: начальную и конечную
        x1 = int(x[0])
        y1 = int(y[i][0])
        x2 = int(x[-1])
        y2 = int(y[i][-1])
            
        cv.line(image,(x1,y1),(x2,y2),(0,255,0), 2)   

drawhoughLinesOnImage(houghLinesImage, max_line) 

# К изображению добавляются разные веса, чтобы создать ощущение смешивания
def blend_images(image, final_image, alpha=0.7, beta=1., gamma=0.):
    return cv.addWeighted(final_image, alpha, image, beta,gamma)

orginalImageWithHoughLines = blend_images(houghLinesImage,image) # и обединяем вместе, используя image blending

# показ 3х картинок 1 первоначальная 2 картинка граней 3 картинка линий + первоначальное изображение
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(20, 20))
ax1.imshow(image)
ax1.set_title('Original Image')
ax1.axis('off')

ax2.imshow(edge, cmap='gray')
ax2.set_title('Edge Image')
ax2.axis('off')

ax3.imshow(orginalImageWithHoughLines, cmap='gray')
ax3.set_title("Original Image with Hough lines")
ax3.axis('off')

plt.imshow(orginalImageWithHoughLines)
```
