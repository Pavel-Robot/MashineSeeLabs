```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
#%matplotlib inline
```

```python
#  
image = cv.imread('road1.png')
print(f"size image:{image.shape}")
plt.imshow(image)
```

```python
#     RGB   
gray_img = cv.cvtColor(image, cv.COLOR_RGB2GRAY)
#plt.imshow(gray_img, cmap='gray')
```

```python
#   
#   c      55
kernel = np.ones((5,5),np.float32)/25
filter_img = cv.filter2D(gray_img, -1, kernel)
#plt.imshow(filter_img, cmap='gray')
```

```python
#   Canny   (,   ,   )
edge = cv.Canny(filter_img, 50, 100)
plt.imshow(edge, cmap='gray')
```

<h3>   (Hough Transform)</h3>
  (     $\rho$  $\theta$).    :

```python
max_angle = 360//1
k = 1 #   
pk, lk = 4, 1 #          ro
print(f":{max_angle} :{max_angle*k} :{1/k} :{max_angle*k*pk*lk*max(image.shape[0],image.shape[1])}")
```

```python
accum = np.zeros((max(image.shape[0],image.shape[1])*pk*lk, max_angle*k), float) # , ro
voices = 200 #   
```

   $\rho=x*cos(\theta)+y*sin(\theta)$,  $x,y$ -        (   ).    

```python
def ro(x, y, theta):
    return x*np.sin(theta*np.pi/180)+y*np.cos(theta*np.pi/180)
print(ro(300, 280, 100))
```

     ,   $\rho$         $O(x_0, y_0)$   $\theta$      *max_angle*.

```python
points = []
tht = [i for i in range(max_angle)]
for x in range(0, len(edge), 1):   
    for y in range(0, len(edge[x]), 1):
        if edge[x][y] != 0: #    (     )
            pts = []
            for theta in range(0, max_angle*k, 1):
                p = ro(x, y, theta/k)*lk
                pts.append(p) #  
                accum[int(p)][theta] += 1
            points.append(pts)
print(max([max(m)/lk for m in points])) #    ro

%matplotlib inline
plt.pcolor(accum, cmap=plt.get_cmap('gist_gray', 11))
plt.colorbar()
plt.show()
```

```python
#     
#     
def func_for_porog(ac, porog):
    s = []
    for i in range(len(ac)):
        for j in range(len(ac[0])):
            if ac[i][j] == 0:
                pass
            else: 
                if ac[i][j] <= porog:
                    ac[i][j] = 0
                else:
                    s.append([ac[i][j], i, j]) #      

    return ac, s
accum, max_line = func_for_porog(accum, voices)
```

     ,      .   $\rho=x*cos(\theta)+y*sin(\theta)$  $y(x)=(\rho-x*sin(\theta))/(cos(\theta))$ .     $y(x)$  ,   c    rho  theta:

```python
def f_y(x, rho, theta):
    return (rho - x*np.cos(theta*np.pi/180))/np.sin(theta*np.pi/180)
    
x = [i for i in range(image.shape[0])]
y = []
for i in range(len(max_line)):
    rho, theta = max_line[i][1], max_line[i][2]
    temp = [f_y(n, rho, theta) for n in x]
    y.append(temp)
```

```python
houghLinesImage = np.zeros_like(image) #   

#  ,   ,   
def drawhoughLinesOnImage(image, lines):
    x = [i for i in range(max(image.shape[0],image.shape[1])*pk*lk)]
    y = []
    for i in range(len(lines)):
        rho, theta = lines[i][1], lines[i][2]
        temp = [f_y(n, rho, theta) for n in x]
        y.append(temp)

        #    :   
        x1 = int(x[0])
        y1 = int(y[i][0])
        x2 = int(x[-1])
        y2 = int(y[i][-1])
            
        cv.line(image,(x1,y1),(x2,y2),(0,255,0), 2)   

drawhoughLinesOnImage(houghLinesImage, max_line) 

#     ,    
def blend_images(image, final_image, alpha=0.7, beta=1., gamma=0.):
    return cv.addWeighted(final_image, alpha, image, beta,gamma)

orginalImageWithHoughLines = blend_images(houghLinesImage,image) #   ,  image blending

#  3  1  2   3   +  
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(20, 20))
ax1.imshow(image)
ax1.set_title('Original Image')
ax1.axis('off')

ax2.imshow(edge, cmap='gray')
ax2.set_title('Edge Image')
ax2.axis('off')

ax3.imshow(orginalImageWithHoughLines, cmap='gray')
ax3.set_title("Original Image with Hough lines")
ax3.axis('off')

plt.imshow(orginalImageWithHoughLines)
```
