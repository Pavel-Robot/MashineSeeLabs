```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

<h2>Определение окружностей на изображении методом Хафа</h2>
Уравнение круга можно описать следующим выражением (если круг находится в центре координат $O(x_0,y_0)$, а координаты точки удаления от этого центра для радиус вектора $r$ это $x,y$ ): <br>

$x^2+y^2=r^2$ <br>

<br>
Если же центр окружности сдвинут относительно начала координат (точка O стала A), то выражение примет вид:

$(x-x_0)^2+(y-y0)^2=r^2$ <br>


<br>
Когда мы попадаем в пиксель x,y на картинке и начинаем перебирать значения x_0, y_0, то узнаем радиус окружности до какого-либо контура. Поэтому нужно по аналогии с нахождением линий, создать аккумулятор трехмерный, где будет x,y,r совпадения учитываться. После вновь производим голосование и таким образом, определим где круги на изображении, а также их параметры.


```python
# Загружаю изображение
image = cv.imread('planets5.png')
print(f"size image:{image.shape}")
plt.imshow(image)
```

```python
# Перевожу цветовое пространство из RGB в оттенки серого
gray_img = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
```

```python
gray_blurred = cv.blur(gray_img, (3, 3)) 
plt.imshow(gray_blurred, cmap='gray')
```

```python
# По методу Canny выделяю границы (изобр, нижний порог яркости, высший порог яркости)
edge = cv.Canny(gray_blurred, 50, 100)
plt.imshow(edge, cmap='gray')
```

3-х мерный массив с размерами изображения $x_{max},y_{max}$ и максимальной длиной $r=\sqrt[2]{x_{max}^2+y_{max}^2}$

```python
max_r = int(1+(image.shape[0]**2+image.shape[1]**2)**0.5)
size = (image.shape[0], image.shape[1], max_r)
accum = np.zeros(size, float) # градусы, ro
voices = 55 # количество голосов 
print(f"size array:{size} and voices:{voices}")
```

```python
# x0,y0 - центр круга, x,y - точка радиуса
def r(x0, y0, x, y):
    return ((x-x0)**2+(y-y0)**2)**0.5
print(r(0, 0, 1, 1))
print(r(1, 1, 2, 2))
```

```python
points = []
tht = [i for i in range(max_r)]

# попадаем в пиксел x0,y0 на картинке
for x0 in range(0, len(edge), 1):
    for y0 in range(0, len(edge[x0]), 1):
            for x in range(0, len(edge), 1):
                for y in range(0, len(edge[x]), 1):
                    if edge[x][y] != 0: # Отсеиваю нулевые значения (где значение пикселя равно черному цвету)
            # начинаем перебирать все возможные точки-радиусы на изображении
            # на этом этапе можно ограничить размер радиуса, если мы определенный объект желаем найти
                        # считаем длину радиуса
                        radius = int(r(x0, y0, x, y))
                        #if radius == 13:
                        # добавляем в акб
                        accum[x0][y0][radius] += 1
```

Работать с таким 4-х мерным акб не очень удобно, поэтому сначала переводим в кортеж вида координат + его значения (x,y,z, accum[x][y][z]) и смотрим, чтобы у найденных окружностей хватало голосов. Формируем список circles в котором есть координаты начала окружности и значение радиуса.

```python
# учитываем порог
circles = []
maximus = [0,0,0,0]
for x in range(0, len(edge), 1):
    for y in range(0, len(edge[x]), 1):
        for z in range(len(accum[x][y])):
            if accum[x][y][z] >= voices:
                if accum[x][y][z] > maximus[3]:
                    maximus = [x, y, z, accum[x][y][z]]
                    circles.append((x,y,z,accum[x][y][z]))
            else:
                accum[x][y][z] = 0
#print(circles)
print(len(circles))
print(circles[-1])
print(max(circles, key=lambda x: x[2]))
circles = [max(circles, key=lambda x: x[3])]
print(circles)
```

Поскольку нам известны координаты и радиус, мы можем изобразить окружность исходя из следующих соображений: <br>
$x = x_0 + r \cdot cos\theta$ <br>
$y = y_0 + r \cdot sin\theta$ <br> 

```python
for i in range(1):
    print(circles)
    cv.circle(image, (circles[i][1], circles[i][0]), circles[i][2], (0, 255, 0), 2) 

# Отображение графика
plt.imshow(image, cmap='gray')
```
