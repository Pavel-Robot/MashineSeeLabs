3-      $x_{max},y_{max}$    $r=\sqrt[2]{x_{max}^2+y_{max}^2}$

```python
max_r = int(1+(image.shape[0]**2+image.shape[1]**2)**0.5)
size = (image.shape[0], image.shape[1], max_r)
accum = np.zeros(size, float) # , ro
voices = 55 #   
print(f"size array:{size} and voices:{voices}")
```

```python
# x0,y0 -  , x,y -  
def r(x0, y0, x, y):
    return ((x-x0)**2+(y-y0)**2)**0.5
print(r(0, 0, 1, 1))
print(r(1, 1, 2, 2))
```

```python
points = []
tht = [i for i in range(max_r)]

#    x0,y0  
for x0 in range(0, len(edge), 1):
    for y0 in range(0, len(edge[x0]), 1):
            for x in range(0, len(edge), 1):
                for y in range(0, len(edge[x]), 1):
                    if edge[x][y] != 0: #    (     )
            #     -  
            #       ,      
                        #   
                        radius = int(r(x0, y0, x, y))
                        #if radius == 13:
                        #   
                        accum[x0][y0][radius] += 1
```

   4-     ,        +   (x,y,z, accum[x][y][z])  ,      .   circles         .

```python
#  
circles = []
maximus = [0,0,0,0]
for x in range(0, len(edge), 1):
    for y in range(0, len(edge[x]), 1):
        for z in range(len(accum[x][y])):
            if accum[x][y][z] >= voices:
                if accum[x][y][z] > maximus[3]:
                    maximus = [x, y, z, accum[x][y][z]]
                    circles.append((x,y,z,accum[x][y][z]))
            else:
                accum[x][y][z] = 0
#print(circles)
print(len(circles))
print(circles[-1])
print(max(circles, key=lambda x: x[2]))
circles = [max(circles, key=lambda x: x[3])]
print(circles)
```
